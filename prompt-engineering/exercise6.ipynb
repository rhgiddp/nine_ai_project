# Prompt Engineering 파트 실습 6 - Prompt Engineering에 대한 현실 및 제약

프롬프트 엔지니어링 이론 및 실습 파트를 마무리하면서 3가지 정도 이야기를 간략하게 해볼 예정

### 1. 프롬프트 엔지니어링
- 모델의 입력 텍스트를 조정해서 LLM 모델의 성능을 추가적으로 끌어올리는 방법
	- 사전훈련 단계에서 습득한 In-Context Learning 능력 활용



### 2. 프롬프트 엔지니어링 장단점 정리
- 추가 학습이 없음에도 성능 개선이 있기 때문에 가성비가 굉장히 좋음
- 더 좋은 모델을 사용하면 프롬프트 엔지니어링 없어도 해결되는 케이스들 다수 존재
	- 앞으로도 더 좋은 모델들은 계속 출시 예정
- 또한 모델이 바뀌면 제너럴한 방법론 말고 대부분이 달라짐


### 3. 프롬프트 엔지니어링의 미래? 대세 여부? 고연봉의 Prompt Engineer?
- **장단점을 명확하게 알고, 수많은 방법론 중 선택 할 수 있는 방법론이라는 인식**
	- 프롬프트 엔지니어링으로 시작하고, 그럼에도 부족하면 파인튜닝 같은 다른 방법론 시도
